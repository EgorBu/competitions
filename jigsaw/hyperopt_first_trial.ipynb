```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials

from fasttext_worker import pipeline, LABEL_COLUMNS
```

```python
space = {
    "train_path": hp.choice("train_path", ["input/train.csv.zip"]),
    "test_path": hp.choice("test_path", ["input/test.csv.zip"]),

    "save_proba": hp.choice("save_proba", ["probas/"]),
    "supervised_thread": hp.choice("supervised_thread", [4]),
    "supervised_label": hp.choice("supervised_label", ["__label__"]),


    "supervised_input": hp.choice("supervised_input", ["output/train_fasttext.txt"]),

    "supervised_output": hp.choice("supervised_output", ["model"]),
    "supervised_lr": hp.uniform("supervised_lr", 0.01, 0.6),
    "supervised_lrUpdateRate": hp.choice("supervised_lrUpdateRate", [2, 5]),
    "supervised_dim": hp.choice("supervised_dim", [20, 40, 80, 150, 200, 300, 400]),
    "supervised_ws": hp.choice("supervised_ws", [2, 3, 4, 5, 7, 10]),
    "supervised_epoch": hp.choice("supervised_epoch", [3, 5, 8, 10, 15, 20]),
    "supervised_minCount": hp.randint("supervised_minCount", 200),
    "supervised_neg": hp.choice("supervised_neg", [3, 5, 10]),
    "supervised_wordNgrams": hp.choice("supervised_wordNgrams", [1, 2, 3, 4]),    
    
    "supervised_minn": hp.choice("supervised_minn", [2, 3, 4]),
    "supervised_maxn": hp.choice("supervised_maxn", [5, 6, 7, 8]),
    "supervised_pretrainedVectors": hp.choice("supervised_pretrainedVectors", [None, "input/crawl-300d-2M.vec"])
}
```

```python
def f(params):
    if params["supervised_pretrainedVectors"] is None:
        del params["supervised_pretrainedVectors"]
    else:
        # not enough memory
        del params["supervised_pretrainedVectors"]
    res = 1 - pipeline(params)
    return {'loss': res, 'status': STATUS_OK}

trials = Trials()
best = fmin(fn=f, space=space, algo=tpe.suggest, max_evals=50, trials=trials)
```
